### Simple Histogram Example

```python
#Histograms show frequency distribution of samples in numerical ranges using bars. 

import matplotlib.pyplot as plt
import numpy as np

# Load sample tempearture values into an array
temperatures = [66, 70, 69, 80, 68, 67, 72, 70, 70, 57, 63, 70, 78, 67, 53, 67, 75, 70, 81, 76, 79, 75, 76, 58, 29]

#Print some general information baout this data using NumPy
print("the mean of temps is : " + str(np.mean(temperatures))) #The mean is simply the average
print("the median of temps is: " + str(np.median(temperatures))) #Median is a midpoint where half data are smaller and half are larger

# Create histogram - this does everything!
plt.figure(figsize=(10, 6))
plt.hist(temperatures, bins=range(min(temperatures), max(temperatures) + 2), edgecolor='black', color='lightgray')

plt.xticks(np.arange(30, max(temperatures) + 5, 5)) #starting tick is 30, add 5 to max temp for last tick to accomodate highest samples, 5 is distance between ticks
plt.title('Temperature Frequency')
plt.xlabel('Temperature')
plt.ylabel('Frequency')
plt.show()
```

### Simple pychart Example

```python
import pandas as pd
#Load synthetic data of students and geography and store in a DataFrame

# Create data as a dictionary with synethetic data
data = {
    'Student_location': [
        'California', 'California', 'California', 'California', 
        'California', 'California', 'California', 'California',
        'Oregon', 'Oregon', 'Oregon', 'Oregon',
        'Washington', 'Washington', 'Washington',
        'Other (US)', 'Other (US)', 'Other (US)',
        'International', 'International'
    ],
    'Student_name': [
        'John Smith', 'Emily Johnson', 'Michael Brown', 'Sarah Davis',
        'David Wilson', 'Jessica Miller', 'James Taylor', 'Emma Anderson',
        'Robert Thomas', 'Ashley Jackson', 'Daniel White', 'Amanda Harris',
        'Christopher Martin', 'Jennifer Thompson', 'Matthew Garcia',
        'Lisa Martinez', 'Andrew Robinson', 'Michelle Clark', 
        'Joshua Rodriguez', 'Karen Lewis'
    ]
}

# Create DataFrame from dictionary
df = pd.DataFrame(data)
display(df.head()) #Display some of the synthetic data

#EXAMPLE 1: Pie chart example (better for understanding sample relative to total)
print() #Insert cosmetic line break
df['Student_location'].value_counts().plot.pie(ylabel='', title='Student Geographic Distribution Pie Chart')
plt.show()

#EXAMPLE 2: Dot plot example (better than pie chart for comparing two categories)
print() #Insert cosmetic line break
percents = df['Student_location'].value_counts(normalize=True) * 100 # Convert to percentages
plt.plot(percents.values, percents.index, 'o') # Plot dots
plt.title('Student Geographic Distribution Dot Plot')
plt.xlabel('Percent')
plt.xlim(0, 50)
plt.grid(axis='x', alpha=0.3)
plt.show()
```

### Boxplot (box-and-whisker plot)

```python
#Box and whisker plots visualize 5 key numbers of the data.  
```

```python
#Create synthetic data of fuel consumption for 32 cars to display on a boxplot
carData = {
    'vehicle_model': [
        'Honda Civic', 'Toyota Camry', 'Ford F-150', 'Chevy Silverado',
        'BMW 3 Series', 'Mercedes C-Class', 'Audi A4',
        'Mazda CX-5', 'Honda CR-V', 'Toyota RAV4', 'Ford Explorer',
        'Jeep Wrangler', 'Chevy Tahoe', 'GMC Sierra', 'Ram 1500',
        'Nissan Altima', 'Hyundai Sonata', 'Kia Optima', 'VW Jetta',
        'Subaru Outback', 'Mazda3', 'Honda Accord', 'Toyota Corolla',
        'Ford Mustang', 'Chevy Camaro', 'Dodge Charger', 'Nissan Rogue',
        'Hyundai Tucson', 'Kia Sportage', 'VW Tiguan', 'Subaru Impreza'
    ],
    'milesPerGallon': [
        32, 28, 20, 19,
        27, 26, 28, 25, 
        28, 27, 21, 17, 
        15, 18, 19, 28, 
        30, 31, 29, 26, 
        30, 30, 33, 22, 
        20, 21, 27, 26, 
        28, 24, 28
    ]
}

# Create DataFrame from dictionary
df = pd.DataFrame(carData)
print(f"\nTotal vehicles: {len(df)}")
```

- Explanation of the box-and-whisker plot: 
   - The lower whisker: maps to the smallest number in the data
   - The upper whisker: maps to the largest number in the data
   - The middle most line: represents the median (e.g. point where half of the data are smaller and half are larger)
   - Bottom of the box: the first quartile (number where 1/4 of data are smaller and 3/4 are larger)
   - Top of the box: the third quartile (point where 3/4 of the data are smaller and 1/4 are larger)

Although box plots convey less information than histograms, they take up less space and therefore are well suited to compare several datasets.

```python
plt.boxplot(df['milesPerGallon'])
plt.ylabel('Miles Per Gallon')
plt.title('Fuel Efficiency box-and-whisker plot of the vehicle fleet')
plt.xlabel('Vehicle fleet')
plt.xticks([]); #Removes the default "1" from x axis
```
