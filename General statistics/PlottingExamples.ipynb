# MatPlot Lib Descriptive Statstics Examples

This notebook demonstrates the basics of plotting and leverages plotting utilities and functions from MatPlotLib. ### Simple Histogram Example

```python
import matplotlib.pyplot as plt
import numpy as np

# Load sample tempearture values into an array
temperatures = [66, 70, 69, 80, 68, 67, 72, 70, 70, 57, 63, 70, 78, 67, 53, 67, 75, 70, 81, 76, 79, 75, 76, 58, 29]

# Create histogram - this does everything!
plt.figure(figsize=(10, 6))
plt.hist(temperatures, bins=10, edgecolor='black', color='lightgray')
plt.xlabel('Temperature')
plt.ylabel('Frequency')
plt.show()
```

### Simple pychart and Dot Plot Example

```python
import pandas as pd
#Load synthetic data of students and geography and store in a DataFrame

# Create data as a dictionary with synethetic data
data = {
    'Student_location': [
        'California', 'California', 'California', 'California', 
        'California', 'California', 'California', 'California',
        'Oregon', 'Oregon', 'Oregon', 'Oregon',
        'Washington', 'Washington', 'Washington',
        'Other (US)', 'Other (US)', 'Other (US)',
        'International', 'International'
    ],
    'Student_name': [
        'John Smith', 'Emily Johnson', 'Michael Brown', 'Sarah Davis',
        'David Wilson', 'Jessica Miller', 'James Taylor', 'Emma Anderson',
        'Robert Thomas', 'Ashley Jackson', 'Daniel White', 'Amanda Harris',
        'Christopher Martin', 'Jennifer Thompson', 'Matthew Garcia',
        'Lisa Martinez', 'Andrew Robinson', 'Michelle Clark', 
        'Joshua Rodriguez', 'Karen Lewis'
    ]
}

# Create DataFrame from dictionary
df = pd.DataFrame(data)
display(df.head()) #Display some of the synthetic data

#EXAMPLE 1: Pie chart example (better for understanding sample relative to total)
print() #Insert cosmetic line break
df['Student_location'].value_counts().plot.pie(ylabel='', title='Student Geographic Distribution Pie Chart')
plt.show()

#EXAMPLE 2: Dot plot example (better than pie chart for comparing two categories)
print() #Insert cosmetic line break
percents = df['Student_location'].value_counts(normalize=True) * 100 # Convert to percentages
plt.plot(percents.values, percents.index, 'o') # Plot dots
plt.title('Student Geographic Distribution Dot Plot')
plt.xlabel('Percent')
plt.xlim(0, 50)
plt.grid(axis='x', alpha=0.3)
plt.show()
```
