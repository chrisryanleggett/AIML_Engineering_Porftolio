```python
import torch

#Create a 1D tensor called a , create a python list and cast it using the tensor constructor
a= torch.tensor([7, 4, 3, 2, 6])

#As with list we can access the elements with an int and square bracket
print(a[:5]) # First 5 elements
print(a) #Print the entire tensor

#Use item() method to convert each tensor element to a Python scalar
for element in a:
    print(element.item(), end=' ')
print()
```

```python
#Use the data attribute dtype to find the type of data stored in the tensor
a = torch.tensor([0, 1, 2, 3, 4])
print("Integer tensor:")
print(a.dtype)
print(a.type())

a = torch.tensor([0.0, 1.0, 2.0, 3.0, 4.0])
print("\nFloat tensor:")
print(a.dtype)
print(a.type())

#Specify the data type of a tensor in the constructor using the parameter dtype
#Effectively a data type override
a = torch.tensor([0, 1, 2, 3, 4], dtype=torch.int32)  # Use integers, not floats
print("\nOverride to int32:")
print(a.dtype)

#Using the FloatTensor method to create a float tensor explicitly
a=torch.FloatTensor([0, 1, 2, 3, 4])
print("\nUsing FloatTensor Method:")
print(a.type())
print(a)

#Convert the tensor type using methods
a=torch.tensor([0,1,2,3,4]) #Tensor of long type
a=a.type(torch.FloatTensor) #Convert to Float using the type method w argument torch.FloatTensor
print("\nType converted from long to float using type():")
print(a.type())
```

```python
#Exploring tensor attributes

#Use size method to return number of elements in a tensor
a = torch.Tensor([0, 1, 2, 3, 4])
print(a.size())

# Determining tensor dimension or rank using ndimension()
print("\nNumDimensions:")
print(a.ndimension())

#Converting 1D tensors to 2D tensors
a = torch.Tensor([1, 2, 3, 4, 5]) 
a_col = a.view(5, 1)  # Convert to 2D tensor using View method, 5 rows 1 col
print("\nOriginal 1D tensor:")
print(a)
print("\nConverted to 2D (5x1):")
print(a_col)
```

```python
#Pytorch tensors can be interconverted to NumPy arrays and vice versa
import numpy as np
import pandas as pd

np_array= np.array([0.0, 1.0, 2.0, 3.0, 4.0])

#convert np array to torch_tensor
torch_tensor = torch.from_numpy(np_array)
print(torch_tensor)
#convert torch_tensor back to np array
np_array_again = torch_tensor.numpy()
print(np_array_again)

pandas_series=pd.Series([0.1, 2, 0.3, 10.1])
#Pandas to torch
pandas_to_torch=torch.from_numpy(pandas_series.values)
print(pandas_to_torch)
```

```python
#Indexing and slicing methods to access a value or set of values in a tensor

c=torch.tensor([20, 1, 2, 3, 4])
print(c[0])

#Change the fifth element of the tensor from 4 to 0
c[4]=0
print(c)

#select elements at index 1 to 4 in tensor c to new torch tensor d, not counting element at last index
print("\nNew tensor d is:")
d = c[1:4] #assign the 2nd and
print(d)
```

## Basic 1D tensor euclidian vector operations

```python
#Vector addition , define tensors u and v of the same type
u = torch.tensor([1, 0])  
v = torch.tensor([0, 1])  
z = u + v
print("\nVector addition result:")
print(z)

#vector multiplication of a scalar
y = torch.tensor([1, 2])  
y = 3 * y
print("\nVector multiplication result:")
print(y)

#Product entry-wise
u = torch.tensor([1, 2])  
v = torch.tensor([3, 2])  
z = u * v 
print("\nElement-wise product result:")
print(z)

#dot product u^T*v using dot
u = torch.tensor([1, 2])  
v = torch.tensor([3, 1])
result = torch.dot(u,v)
print("\ndot product result:")
print(result)
print("\n")

#adding constant to vector
u=torch.tensor([1,2,3,-1])
z= u+1 #broadcasting
print("\nAfter adding constant 1 to tensor [1, 2, 3, -1]:")
print(z)

#apply functions to tensors
print("\n")
mean_a = torch.tensor([1, -1, 1, -1]).float().mean()
print("Mean of tensor [1, -1, 1, -1] is:", mean_a) #balanced tensor where mean is exactly 0.
```

## Differentiation and derivatives in PyTorch Examples

```python
# Import SymPy (symbolic math) and library to show basic latest equations in notebooks
from sympy import *
from IPython.display import display

print("We are going to take the derivative of the following equation:")
x, y = symbols('x y')
eq = Eq(y, x**2)
display(eq)
```

```python
import torch 
import matplotlib.pylab as plt

#Since we are taking the derivative of y with respect to x where x = 2
# Create a tensor x
x = torch.tensor(2.0, requires_grad = True)

#Create the equation 
y = x ** 2

y.backward() #take derivative of y
print("The derivative of y at x = 2 : ", x.grad)
```

The preceding lines perform the following operation:

$\frac{\mathrm{dy(x)}}{\mathrm{dx}}=2x$


$\frac{\mathrm{dy(x=2)}}{\mathrm{dx}}=2(2)=4$


```python
print("Example: Take the derivative of the following equation where x = 1:")
x, y = symbols('x y')
eq = Eq(y, 2*x**3 + x)
display(eq)
```

```python
#where x = 1
x = torch.tensor(1.0, requires_grad = True)

#Create the equation
y = 2*x**3 + x

#Take the derivative of y with respect to x
y.backward()
print("The derivative of y at x = 1 is: ", x.grad)
```

## Partial derivatives in PyTorch

- Partial derivatives involve functions of multiple variables.  The derivatives we epxlored previously are functions of a single variable.
- The below section of the notebook showcases how we solve partial derivatives in PyTorch

```python
print("Take the partial derivative of the following equation where u = 1 and v = 2")
u, v = symbols('u v')
f = Function('f')
eq = Eq(f(u, v), v*u + u**2)
display(eq)
```

```python
import numpy as np
import matplotlib.pyplot as plt
from mpl_toolkits.mplot3d import Axes3D

# SOLVE THE SYSTEM
#Create the tensors u and v at u=1 and v =2 
u = torch.tensor(1.0,requires_grad=True)
v = torch.tensor(2.0,requires_grad=True)

#Create the equation for the function
f= u * v + u **2
f.backward() # Compute gradients (e.g. partial derivatives) of f with respect to u and v

#calculate the derivatives with respect to u and v
print("Partial derivative w/ respect to u: ", u.grad)
print("Partial derivative w/ respect to v: ", v.grad)

#PLOT THE SYSTEM
# df/du - accounted for in steepness when moving in the u direction 
# df/dv = accounted for in the steepness when moving in the v direction 
# color = solution at the intersection w/ mesh at respective` values of u and v

# Create mesh grid
u_grid = np.linspace(-3, 3, 50)
v_grid = np.linspace(-2, 4, 50)
U, V = np.meshgrid(u_grid, v_grid)

# Calculate function values
F = U * V + U**2

# Create 3D surface plot
fig = plt.figure(figsize=(10, 8))
ax = fig.add_subplot(111, projection='3d')

# Plot surface with more sensitive color gradient
surf = ax.plot_surface(U, V, F, cmap='plasma', alpha=0.9, 
                     vmin=np.percentile(F, 5), vmax=np.percentile(F, 95))

# Mark the point (u=1, v=2)
ax.scatter([1], [2], [3], color='red', s=100, label='(u=1, v=2, f=3)')

# Labels and title
ax.set_xlabel('u')
ax.set_ylabel('v') 
ax.set_zlabel('f(u,v)')
ax.set_title('f(u,v) = uv + u²')
plt.colorbar(surf, shrink=0.5)
ax.legend()
plt.show()
```
